#### Create a new column basis count option

```r
flights %>%
  mutate(long_flight = (air_time >= 6 * 60)) %>%
  View()
```

```r
## count the number of long flights
flights %>%
    count(long_flight = air_time >= 6*60)
```

#### create a new column basis group

```r
flights %>%
  group_by(date = make_date(year, month, day)) %>%
  summarise(flights_n = n(), air_time_mean = mean(air_time, na.rm = TRUE)) %>%
  ungroup()
```

#### Data column creation

```r
flights %>%
  select(year,month,day) %>%
  mutate(date = make_date(year,month,day))
```

### number parsing

```r
## If we want to extract only numbers then parse_option

numbers_1 <- tibble(number = c("#%$%@$%123454", "Numbers668", "Spider 367"))
numbers_1 %>% mutate(number = parse_number(number))
```

#### str_replace_all to find and replace multiple options at once

```r
flights %>%
  mutate(origin = str_replace_all(origin, c(
    "^EWR$" = "Newark International",    "^JFK$" = "John F. Kennedy International"
  ))) %>%
  count(origin)
```

#### Select columns with starts_with and ends_with

```r
flights %>%
  select(starts_with("dep_"))
flights %>%
  select(ends_with("hour"))
flights %>%
  select(contains("hour"))
```

##### case_when to create when conditions are met

```r
flights %>%
  mutate(origin = case_when(
    (origin == "EWR") & dep_delay > 20 ~ "Newark International Airport - DELAYED",
    (origin == "EWR") & dep_delay <= 20 ~ "Newark International Airport - ON TIME DEPARTURE",
  )) %>%
  count(origin)
```
